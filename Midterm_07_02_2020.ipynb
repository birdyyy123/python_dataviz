# ข้อ 1
สร้าง function รับ input เป็น list 2 ตัว ที่มีขนาดเท่ากัน โดย list แรกเป็น list ของตัวเลข list ที่ สองเป็น list ของ text
- ถ้าค่าใน list แรกเป็น เลขคี่ ให้ค่าใน dict เป็น output มี key เป็น่าจาก list ตัวที่ 1 และ value จาก list ตัวที่ 2
- ถ้าค่าใน list แรกเป็น เลขคู่ ให้ค่าใน dict เป็น output มี key เป็น่าจาก list ตัวที่ 2 และ value จาก list ตัวที่ 1
- output ของ function คือ dict ที่สร้างขึ้น

```python
def smartDict(num,text):
    f ={}
    for i in range(len(num)):
        if num[i]%2 ==1:
            f[num[i]] = text[i]
        else:
            f[text[i]] = num[i]
    return(f)    
```

```python
smartDict([1,2,3],['apple','banana','orange'])
```

```python
smartDict([7,1,4,6],['A','B','E','Zzz'])
```

-2.1 กรดอะมิโนใด (A,T,G,C) มีปริมาณมากที่สุดและมีเท่าไหร่

```python
import pandas as pd
```

```python
DNA_df = pd.read_csv('Wuhan_nCoV_GNome.txt',header=None)
```

```python
def str1(seq):
    str23 = []
    for i in range(len(seq)):
        str23.append(seq[i])
    return str23
```

```python
liststr1 = []
for i in range(len(DNA_df)):
    liststr1 += str1(DNA_df.iloc[i][0])
```

```python
unique_1char=set(liststr1)
```

```python
dict_1char = {}
for i in unique_1char:
    num = 0
    for j in liststr1:
        if i == j:
            num += 1
    dict_1char[i] = num
```

```python
dict_1char
```

```python
Max_V = -100
Maxkey = None
for i in dict_1char.keys():
    if dict_1char[i] > Max_V:
        Max_V = dict_1char[i]
        Maxkey = i
```

```python
print(f'กรดอะมิโนที่มากที่สุดคือ {Maxkey} จำนวน {Max_V} ตัว')
```

-2.2 จงหา sequence ที่มีขนาด 3 และ 4 ที่เกิดขึ้นซ้ำมากที่สุด  

```python
def str4(seq):
    str23 = []
    for i in range(len(seq)-3):
        str23.append(seq[i:i+4])
    return str23
```

```python
def str3(seq):
    str23 = []
    for i in range(len(seq)-2):
        str23.append(seq[i:i+3])
    return str23
```

```python
unique_4char=set(liststr4)
unique_3char=set(liststr3)
```

```python
dict_4char = {}
for i in unique_4char:
    num = 0
    for j in liststr4:
        if i == j:
            num += 1
    dict_4char[i] = num
```

```python
dict_3char = {}
for i in unique_3char:
    num = 0
    for j in liststr3:
        if i == j:
            num += 1
    dict_3char[i] = num
```

```python
MaxVstr3 = -100
Maxkey3 = None
for i in dict_3char.keys():
    if dict_3char[i] > MaxVstr3:
        MaxVstr3 = dict_3char[i]
        Maxkey3 = i
```

3.1 สร้างตารางใหม่ที่มี 2 columns โดย columns แรกชื่อว่าภาษาอีสาน และมีค่าใน columns เป็นค่า จาก columns ภาษาอีสานเดิมแต่เปลี่ยน ค่าที่เป็น '/'เป็น ',' และ
    columns ที่สองชื่อว่าภาษาไทย และมีค่าใน columns เป็นค่า จาก columns ภาษาไทยเดิมแต่เปลี่ยน ค่าที่เป็น '/'เป็น ','

```python
import pandas as pd
```

```python
L = ['ภาษาอีสาน' , 'ภาษาไทย']
Data = pd.DataFrame.from_records([],columns = L)
for i in range(32):
    df = pd.read_excel('data.xlsx', sheet_name = i, usecols = 'B:C', skipfooter = 1)
    Data = pd.concat([Data,df], axis = 0, ignore_index = True)
```

```python
Data
```

```python
def replacestr(col1):
    co1 = col1.replace("/", ",")
    return co1
```

```python
list_1 = []
list_2 = []
for i in range(len(Data)):
    list_1.append(replacestr(Data.iloc[i]['ภาษาอีสาน']))
    list_2.append(replacestr(Data.iloc[i]['ภาษาไทย']))
```

```python
T1 = [('ภาษาอีสาน', list_1),('ภาษาไทย', list_2)]
Table1 = pd.DataFrame.from_items(T1)
Table1
```

-3.2 สร้างตาราง dictionary โดยตารางประกอบด้วยสอง columns คือ ภาษาอีสาน และ ภาษาไทย โดยค่าใน column ภาษาอีสาน คือคำในภาษาอีสาน และ column ภาษาไทย คือคำแปลเป็นภาษากลางของภาษาอีสาน ใน column แรก

```python
def splitstr(col1):
    co1 = col1.split('/')
    return co1
```

```python
list_11 = []
list_22 = []
for i in range(len(Data)):
    if len(splitstr(Data.iloc[i]['ภาษาอีสาน'])) == len(splitstr(Data.iloc[i]['ภาษาไทย'])): # ตัดก่อนแล้วค่อยดู
        list_11 += splitstr(Data.iloc[i]['ภาษาอีสาน'])
        list_22 += splitstr(Data.iloc[i]['ภาษาไทย'])
```

```python
T2 = {'ภาษาอีสาน': list_11,'ภาษาไทย':list_22 }
Table2 = pd.DataFrame(T2)
Table2
```
