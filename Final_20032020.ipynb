```python
import pandas as pd # import libary
colo = 'C://Users//Minny//Downloads//2019_nCoV_data.csv'
data = pd.read_csv(colo)
from matplotlib import pyplot as plt
```

```python
data.isnull().any()
```

```python
data['Province/State']=data['Province/State'].fillna('Other')
```

```python
data
```

```python
data2 = data[data['Country']!='Mainland China'] #ตัด Mainland China ใน Country columns ออก
data3 = data2[data2['Country']!='Brazil'] #ตัด Brazil ใน Country columns ออก
data4 = data3[data3['Country']!='Mexico'] #ตัด Mexico ใน Country columns ออก
data5 = data4[data4['Country']!='China'] #ตัด China ใน Country columns ออก
data6 = data5[data5['Country']!='Others'] #ตัด Others ใน Country columns ออก
```

```python
h= data6.groupby('Country').max().reset_index()
h
```

```python
plt.bar(h.iloc[:,0],h.iloc[:,5],color = 'r')
plt.title(u'จำนวนผู้ติดเชื้อของแต่ละประเทศ',fontname='Tahoma',fontsize='13')
plt.xlabel('Country',fontname='Tahoma',fontsize='11')
plt.ylabel('Confirmed')
```

```python
plt.bar(h.iloc[:,0],h.iloc[:,6],color = 'g')
plt.title(u'จำนวนผู้ที่ตายของแต่ละประเทศ',fontname='Tahoma',fontsize='13')
plt.xlabel('Country',fontname='Tahoma',fontsize='11')
plt.ylabel('Deaths')
```

```python
plt.bar(h.iloc[:,0],h.iloc[:,7],color = 'b')
plt.title(u'จำนวนผู้ที่รักษาหายของแต่ละประเทศ',fontname='Tahoma',fontsize='13')
plt.xlabel('Country',fontname='Tahoma',fontsize='11')
plt.ylabel('Recovered')
```

```python
labels=['Confirmed','Deaths','Recovered']
```

```python
plt.pie([data['Confirmed'].sum(),data['Deaths'].sum(),data['Recovered'].sum()],labels=labels)
plt.title(u'สัดส่วนคนติดเชื่อ ตาย และรักษาหาย',fontname='Tahoma',fontsize='13')
```

```python
line_data= data.groupby('Date').sum().reset_index()
```

```python
plt.plot(line_data.iloc[:,0],line_data.iloc[:,5], marker='*',color = 'r')
plt.plot(line_data.iloc[:,0],line_data.iloc[:,6], marker='>',color = 'g')
plt.plot(line_data.iloc[:,0],line_data.iloc[:,7], marker='.',color = 'b')
plt.title(u'แนวโน้วคนติดเชื่อ ตาย และรักษาหาย',fontname='Tahoma',fontsize='13')
```
