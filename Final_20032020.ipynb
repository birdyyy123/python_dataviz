```python
import pandas as pd # import libary
colo = 'C://Users//Minny//Downloads//2019_nCoV_data.csv'
data = pd.read_csv(colo)
from matplotlib import pyplot as plt
```

```python
data.isnull().any()
```

```python
data['Province/State']=data['Province/State'].fillna('Other')
```

```python
data
```

```python
data2 = data[data['Country']!='Mainland China']
data3 = data2[data2['Country']!='Brazil']
data4 = data3[data3['Country']!='Mexico']
data5 = data4[data4['Country']!='China']
data6 = data5[data5['Country']!='Others']
```

```python
h= data6.groupby('Country').max().reset_index()
h
```

```python
plt.bar(h.iloc[:,0],h.iloc[:,5],color = 'r')
plt.title(u'จำนวนผู้ติดเชื้อของแต่ละประเทศ',fontname='Tahoma',fontsize='13')
plt.xlabel('Country',fontname='Tahoma',fontsize='11')
plt.ylabel('Confirmed')
```

```python
plt.bar(h.iloc[:,0],h.iloc[:,6],color = 'g')
plt.title(u'จำนวนผู้ที่ตายของแต่ละประเทศ',fontname='Tahoma',fontsize='13')
plt.xlabel('Country',fontname='Tahoma',fontsize='11')
plt.ylabel('Deaths')
```

```python
plt.bar(h.iloc[:,0],h.iloc[:,7],color = 'b')
plt.title(u'จำนวนผู้ที่รักษาหายของแต่ละประเทศ',fontname='Tahoma',fontsize='13')
plt.xlabel('Country',fontname='Tahoma',fontsize='11')
plt.ylabel('Recovered')
```
